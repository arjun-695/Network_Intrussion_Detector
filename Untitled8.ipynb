```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim.lr_scheduler import OneCycleLR
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from collections import Counter
from tqdm import tqdm
import os
import warnings
warnings.filterwarnings('ignore')

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {DEVICE}")
```

```python
import pandas as pd, numpy as np


data = pd.read_csv("/kaggle/input/nslkdd/KDDTrain+_20Percent.txt", header=None)
data.head()
```

```python
COLUMNS = [
    'duration','protocol_type','service','flag',
    'src_bytes','dst_bytes','land','wrong_fragment','urgent',
    'hot','num_failed_logins','logged_in','num_compromised',
    'root_shell','su_attempted','num_root','num_file_creations',
    'num_shells','num_access_files','num_outbound_cmds',
    'is_host_login','is_guest_login','count','srv_count',
    'serror_rate','srv_serror_rate','rerror_rate','srv_rerror_rate',
    'same_srv_rate','diff_srv_rate','srv_diff_host_rate',
    'dst_host_count','dst_host_srv_count','dst_host_same_srv_rate',
    'dst_host_diff_srv_rate','dst_host_same_src_port_rate',
    'dst_host_srv_diff_host_rate','dst_host_serror_rate',
    'dst_host_srv_serror_rate','dst_host_rerror_rate',
    'dst_host_srv_rerror_rate','label','difficulty'
]

def load_nslkdd(path):
    # Read the CSV and assign the proper column names immediately
    df = pd.read_csv(path, header=None, names=COLUMNS)
    # Drop the 'difficulty' score as it's not a real network feature
    df.drop(columns=['difficulty'], inplace=True)
    return df

train_df = load_nslkdd("/kaggle/input/nslkdd/KDDTrain+.txt")
test_df  = load_nslkdd("/kaggle/input/nslkdd/KDDTest+.txt")

print(f"Training samples: {len(train_df):,}")
print(train_df['label'].value_counts())
print("\nTraining label distribution:")
print(train_df['label'].value_counts())
train_df.head()
```

```python
DOS_ATTACKS = ['back', 'land', 'neptune', 'pod', 'smurf', 'teardrop',
               'apache2', 'udpstrom', 'processtable', 'worm']
PROBE_ATTACKS = ['ipsweep', 'nmap', 'portsweep', 'satan','mscan','saint']
R2L_ATTACKS   = ['ftp_write','guess_passwd','imap','multihop','phf',
                  'spy','warezclient','warezmaster','sendmail',
                  'named','snmpgetattack','snmpguess','xlock',
                  'xsnoop','httptunnel']

U2R_ATTACKS   = ['buffer_overflow','loadmodule','perl','rootkit',
                  'sqlattack','xterm','ps']

def map_label(label):
  if label == 'normal': return 0
  if label in DOS_ATTACKS: return 1
  if label in PROBE_ATTACKS: return 2
  if label in R2L_ATTACKS: return 3
  if label in U2R_ATTACKS: return 4
  return 1 #unknown attack -> DoS (conservative fallback)
```

```python
train_df['target'] = train_df['label'].apply(map_label)
test_df['target'] = test_df['label'].apply(map_label)
CLASS_NAMES = ['Normal', 'DoS', 'Probe', 'R2L', 'U2R']
print("Training class distribution:")
for i, name in enumerate(CLASS_NAMES):
    count = (train_df['target'] == i).sum()
    print(f"  {name}: {count:,}")
```

# Feature Engineering

step 1: encode categorial features

```python
CAT_COLS = ['protocol_type', 'service', 'flag']

train_enc = pd.get_dummies(train_df, columns=CAT_COLS)
test_enc = pd.get_dummies(test_df, columns=CAT_COLS)

#align columns
train_enc, test_enc = train_enc.align(
    test_enc, join='left', axis=1, fill_value=0)

print(f'Features after encoding: {train_enc.shape[1]}')
```

step 2: drop label columns, isolate features

```python

DROP = ['label', 'target']
FEAT_COLS = [c for c in train_enc.columns if c not in DROP]

X_train_raw = train_enc[FEAT_COLS].values.astype(np.float32)
y_train     = train_enc['target'].values.astype(np.int64)
X_test_raw  = test_enc[FEAT_COLS].values.astype(np.float32)
y_test      = test_enc['target'].values.astype(np.int64)
```

step-3: scaling features

```python

NUM_COLS = [c for c in FEAT_COLS
            if not any(c.startswith(cat) for cat in CAT_COLS)]

scaler = StandardScaler()

num_idx = [FEAT_COLS.index(c) for c in NUM_COLS]

X_train_raw[:, num_idx] = scaler.fit_transform(
    X_train_raw[:, num_idx])
X_test_raw[:, num_idx]  = scaler.transform(
    X_test_raw[:, num_idx])
```

step 4: engineered interaction features

```python
def compute_interaction_features(df):
  '''
  adding 4 features that capture attack patterns
  invisible in individual column
  '''

  #Ration Bytes send v/s Received (asymmetry = Dos sign)
  src = df['src_bytes'].values.reshape(-1,1).astype(np.float32)
  dst = df['dst_bytes'].values.reshape(-1,1).astype(np.float32) + 1
  byte_ratio = src / dst

  #Error Rate spike (Probe signature)
  err = (df['serror_rate'] * df['serror_rate']).values.reshape(-1,1)

  # Connection density spike (DoS Signature)
  density = (df['count'] * df['same_srv_rate']).values.reshape(-1,1)

  # Guest / failed login flag combo (R2L/U2R signature)
  priv = (df['is_guest_login'] + df['num_failed_logins']).values.reshape(-1,1)

  return np.hstack([byte_ratio, err, density, priv]).astype(np.float32)

X_train = compute_interaction_features(X_train_raw, train_df)
X_test  = compute_interaction_features(X_test_raw,  test_df)

# Scale interaction features with a SEPARATE scaler
extra_scaler = StandardScaler()
X_train= extra_scaler.fit_transform(X_train)
X_test = extra_scaler.transform(X_test)

X_train = np.hstack([X_train_raw, X_train])
X_test  = np.hstack([X_test_raw,  X_test])

N_FEATURES = X_train.shape[1]
print(f'Final feature count: {N_FEATURES}')
```

```python
# SMOTE for minority class balancing

print(f"Before SMOTE: {Counter(y_train)}")

# Determine which classes need oversampling

target_counts = Counter(y_train)
majority_count = max(target_counts.values())
min_target = majority_count // 10

sampling_strategy = {}
for cls, count in target_counts.items():
  if count < min_target:
    sampling_strategy[cls] = min_target

print(f"SMOTE strategy: {sampling_strategy}")

sm = SMOTE(random_state=42, k_neighbors=3, sampling_strategy=sampling_strategy)
X_train_balanced, y_train_balanced = sm.fit_resample(X_train, y_train)
print(f"After SMOTE:  {Counter(y_train_balanced)}")

X_train_balanced, y_train_balanced = shuffle(
    X_train_balanced, y_train_balanced, random_state=42)
```

# Sequence creation for LSTM inputs

```python
SEQ_LEN = 30 # Looks at last 30 connection records

def make_sequences(X, y, seq_len=SEQ_LEN):
  '''
  sliding window over the connection log.
  Label = class of the Last record in the window
  '''

  Xs, ys = [], []
  for i in range(len(X) - seq_len+1):
    Xs.append(X[i:i+seq_len])
    ys.append(y[i+seq_len-1])
  return (np.array(Xs, dtype = np.float32),np.array(ys, dtype = np.int64))



# X_train, y_train = shuffle(X_train, y_train, random_state=42)

X_train_seq, y_train_seq = make_sequences(X_train, y_train)
X_test_seq, y_test_seq = make_sequences(X_test, y_test)

# Verify class distribution

# print(Counter(y_tr_seq))

print(f'Training sequences: {X_train_seq.shape[0]}')
print(f'Test sequences: {X_test_seq.shape[0]}')
print(f'Sequence shape:     {X_train_seq.shape}')
print(f'\nSequence class distribution:')
for cls, count in sorted(Counter(y_train_seq).items()):
    print(f'  {CLASS_NAMES[cls]}: {count:,}')


class IDSDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.FloatTensor(X)
        self.y = torch.LongTensor(y)
    def __len__(self):  return len(self.y)
    def __getitem__(self, i): return self.X[i], self.y[i]


train_ds = IDSDataset(X_train_seq, y_train_seq)
test_ds  = IDSDataset(X_test_seq, y_test_seq)

train_loader = DataLoader(train_ds, batch_size=512,
                          shuffle=True,  num_workers=2)
test_loader  = DataLoader(test_ds,  batch_size=512,
                          shuffle=False, num_workers=2)
print(f"Train batches: {len(train_loader)}")
print(f"Test batches:  {len(test_loader)}")

```

Stacked Bidirectional Lstm architecture

```python
class CNN_BiLSTM_MHA_IDs(nn.Module):
  def __init__(self, n_features , hidden = 128, n_layers= 2, n_classes=5, dropout=0.3):

    super().__init__()

    # Input Projection + normalization
    self.input_proj = nn.Sequential(
        nn.Linear(n_features, hidden),
        nn.LayerNorm(hidden),
        nn.ReLU(),
        nn.Dropout(dropout)

    )

    #1D CNN for local temporal pattern extraction
    self.cnn = nn.Sequential(
        nn.Conv1d(hidden, hidden, kernel_size = 3, padding = 1),
        nn.BatchNorm1d(hidden),
        nn.ReLU(),
        nn.Conv1d(hidden,hidden, kernel_size = 3, padding = 1),
        nn.BatchNorm1d(hidden),
        nn.ReLU(),
    )

    #Stacked Bidirectional LSTM
    self.lstm = nn.LSTM(
            input_size= hidden,
            hidden_size = hidden,
            num_layers = n_layers,
            batch_first= True,
            bidirectional = True, #forward + backward
            dropout = dropout
        )

    #Multi-Head Self Attention(4 Heads)
    self.mha = nn.MultiheadAttention(
        hidden * 2,
        num_heads = 4,
        dropout = 0.1, batch_first = True)

    # Attention pooling over timesteps
    self.attn = nn.LayerNorm(hidden * 2)

    #Learnabel attention aggregation
    self.attn = nn.Linear(hidden * 2,1)

    # Classification head
    self. classifier = nn.Sequential(
        nn.Linear(hidden * 2, 128),
        nn.ReLU(),
        nn.Dropout(dropout),
        nn.Linear(128, n_classes),
        nn.ReLU(),
        nn.Dropout(dropout),
        nn.Linear(128, n_classes)
    )

  def forward(self, x):
    # x: [B, T, F]
    x = self.input_proj(x)  # [B, T, hidden]

    #CNN: extracts local patterns along time axis
    x = x.permute(0,2,1)
    x = self.cnn(x)
    x = x.permute(0,2,1)

    # BiLSTM: capture long range temporal dependencies
    out, _ = self.lstm(x)    # [B, T, hidden*2

    # Multi-head Self-Attention + residula
    attn_out, _= self.mha(out, out, out)
    out = self.attn_norm(out + attn_out)

    #Learnable attention pooling over time
    attn_w = torch.softmax( self.attn(out), dim= 1)    # [B, T, 1]
    context = (out * attn_w).sum(1)     # [B, hidden * 2
    return self.classifier(context) #[B,5]
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
model = CNN_BiLSTM_MHA_IDs(n_features = N_FEATURES).to(DEVICE)

total = sum(p.numel() for p in model. parameters())
print(f'Parameters: {total: ,}') #
print(model)
```

```python
# from torch.optim.lr_scheduler import OneCycleLR
# from tqdm import tqdm
# import os


class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super().__init__()
        self.alpha = alpha     # class weight tensor
        self.gamma = gamma     # focusing parameter
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(
            inputs, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.reduction == 'mean':
            return focal_loss.mean()
        return focal_loss.sum()

# 1. Count occurrences in the sequence labels, forcing an array of exactly 5
# (We use y_train_seq because those are the actual labels the model sees)
class_counts = np.bincount(y_train_seq, minlength=5)

# 2. Prevent division by zero (replaces any 0 counts with 1)
class_counts = np.maximum(class_counts, 1)

# 3. Calculate 'balanced' weights manually
total_samples = len(y_train_seq)
class_weights = total_samples / (5.0 * class_counts)
class_weights = np.clip(class_weights, 0.5, 5.0)

# 4. Convert to Tensor
weights_tensor = torch.FloatTensor(class_weights).to(DEVICE)



print(f"Sequence Class counts: {class_counts}")
print(f"Class weights applied: {class_weights}")

# 5. Pass into the loss function
criterion = FocalLoss(alpha=weights_tensor, gamma=2.0)
EPOCHS = 50
LR = 3e-4

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=LR,
    weight_decay=1e-4
)

scheduler = OneCycleLR(
    optimizer,
    max_lr = LR,
    steps_per_epoch = len(train_loader),
    epochs = EPOCHS,
    pct_start = 0.1  #10% warmup, then cosine decay
)

# Use a gentler scheduler — ReduceLROnPlateau
# scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
#     optimizer,
#     mode='max',        # maximize val accuracy
#     factor=0.5,        # halve LR when plateau hit
#     patience=3,        # wait 3 epochs before reducing
# )

os.makedirs('/content/drive/MyDrive/checkpoints', exist_ok = True)
best_val_acc = 0.0
history = {'train_acc': [], 'val_acc': [], 'loss': [], 'lr': []}

for epoch in range(1, EPOCHS+1):
  model.train()
  total_loss, correct, total = 0, 0, 0

  for xb, yb in tqdm(train_loader, desc = f'Ep {epoch:02d}', leave = False):
    xb, yb = xb.to(DEVICE), yb.to(DEVICE)
    logits = model(xb)
    loss = criterion(logits, yb)


    optimizer.zero_grad()
    loss.backward()
    nn.utils.clip_grad_norm_(model.parameters(), 1.0)
    optimizer.step()
    # scheduler.step()

    total_loss += loss.item()
    preds = logits.argmax(dim = 1)
    correct += (preds == yb).sum().item()
    total += yb.size(0)

  train_acc = correct / total

  model.eval()
  val_correct, val_total = 0,0
  with torch.no_grad():
    for xb, yb in test_loader:
      xb, yb = xb.to(DEVICE), yb.to(DEVICE)
      preds = model(xb).argmax(dim = 1)
      val_correct += (preds == yb).sum(). item()
      val_total += yb.size(0)

  val_acc = val_correct / val_total
  current_lr = optimizer.param_groups[0]['lr']

  history['train_acc'].append(train_acc)
  history['val_acc'].append(val_acc)
  history['loss'].append(total_loss / len(train_loader))
  history['lr'].append(current_lr)

  print(f'Epoch {epoch:02d} | '
            f'Loss: {total_loss/len(train_loader):.4f} | '
            f'Train Acc: {train_acc:.4f} | '
            f'Val Acc: {val_acc:.4f} | '
            f'LR: {current_lr:.2e}')

  if val_acc > best_val_acc:
          best_val_acc = val_acc
          torch.save(model.state_dict(), '/content/drive/MyDrive/checkpoints/best_ids.pth')
          print('  Saved best checkpoint')
print(f'\n{"="*60}')
print(f'Best Validation Accuracy: {best_val_acc:.4f}')
print(f'{"="*60}')
```
